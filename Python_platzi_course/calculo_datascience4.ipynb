```python
import matplotlib.pyplot as plt 
import numpy as np 
def g(x):
    return x*2

def f(x):
    return np.sin(x)

x = np.linspace(-10, 10 , num = 1000) #Genera un vector
fog = f(g(x)) #Función compuesta
plt.plot(x, fog)
plt.show()
```

```python
from matplotlib import cm #Color management
import matplotlib.pyplot as plt 
import numpy as np

#Función de dos variables
def f(x, y):
    return np.sin(x) + 2*np.cos(y)
res = 100

x = np.linspace(-4, 4, num = res)
y = np.linspace(-4, 4, num = res)

x, y = np.meshgrid(x, y)

z = f(x, y)

#Le indicamos al subplot que tipo de gráfica querémos
fig, ax = plt.subplots(subplot_kw={"projection":"3d"}) 

surf = ax.plot_surface(x , y , z, cmap = cm.cool)
fig.colorbar(surf) #Le pone la barrita al lado

fig2, ax2 = plt.subplots()
#Realiza nuestra primera curva de nivel
level_map = np.linspace(np.min(z), np.max(z), num = res)
#Realiza todas las curvas.
cp = ax2.contour(x, y, z, levels = level_map, cmap = cm.cool)
```
